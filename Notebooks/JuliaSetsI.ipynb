```python
from ipywidgets import widgets, interactive
from IPython.display import display
from random import randint
```

```python
def func(z):
    return z*z-0.75

def invfunct(z):
    a=randint(0,1)
    if a==0:
        return (z-0.6-0.1j)**0.5
    else:
        return -(z-0.6-0.1j)**0.5
```

```python
z=1+1j
l=[z]
for i in range(200000):
    z=invfunct(z)
    l=l+[z]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def plot_list (list_complex):
    width = 10
    height = 10
    plt.figure(figsize=(width, height))
    plt.plot([z.real for z in list_complex], [z.imag for z in list_complex],'bo', markersize=0.1)
    plt.axis([-2,2,-2,2])


def plot_list_param (list_complex,a,b,c,d):
    width = 10
    height = 10
    plt.figure(figsize=(width, height))
    plt.plot([z.real for z in list_complex], [z.imag for z in list_complex],'bo', markersize=0.1)
    plt.axis([a,b,c,d])
    plt.savefig("hla")
```

```python
def f(c,d):
    def invfunct_par(z,c,d):
        a=randint(0,1)
        if a==0:
            return (z+c+d*1j)**0.5
        else:
            return -(z+c+d*1j)**0.5
    z=1+1j
    l=[z]
    for i in range(10000):
        z=invfunct_par(z,c,d)
        l=l+[z]
    plot_list(l)
```

```python

a=interactive(f,c=(-1,1,0.01),d=(-1,1,0.01))
display(a)
```
