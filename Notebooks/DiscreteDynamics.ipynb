```python
import matplotlib.pyplot as plt
#from scipy.misc import imread

%matplotlib inline

def plot_list (list_complex):
    width = 10
    height = 10
    plt.figure(figsize=(width, height))
    plt.plot([z.real for z in list_complex], [z.imag for z in list_complex],'ro')
    plt.axis([-4.1, 4.1, -4.1, 4.1])
    
def plot_list_save_variation (list_complex,i):
    width = 10
    height = 10
    #im=plt.imread("Julia_Sets/julia2.png")
    plt.figure(figsize=(width, height))
    #plt.imshow(im,zorder=0, extent=[-2, 2, -1.5, 1.5])
    plt.ylabel('Im(z)')
    plt.xlabel('Re(z)')
    plt.title('f(z)=z*z-z+2')
    plt.plot([z.real for z in list_complex[:2]], [z.imag for z in list_complex[:2]],'ro')
    plt.plot([z.real for z in list_complex[2:4]], [z.imag for z in list_complex[2:4]],'bo')
    plt.plot([z.real for z in list_complex[4:]], [z.imag for z in list_complex[4:]],'go')
    plt.axis([-2, 2, -1.5, 1.5])
    plt.savefig("figure"+str(i)+".png")

def dynamical(z):
    return z*z-z+2
    
```

```python
#Dynamics of z^2-z+2

z1= 0.866025
z2= -0.866025

z3=0.87
z4=-0.85
z5=0.85
z6=-0.87

z5=-0.5

l=[z1,z2,z3,z4,z5]
for i in range(7):
    l=[dynamical(z) for z in l]
    plot_list_save_variation(l,i)
    
```

```python
#Dynamics of z^2-z+2

z1= 1+1j
z2= 1-1j

z3=1.01+1.01j
z4=1.01-0.99j

z5=(2**0.5)*1j

l=[z1,z2,z3,z4,z5]
for i in range(7):
    l=[dynamical(z) for z in l]
    plot_list_save_variation(l,i)
    
```
