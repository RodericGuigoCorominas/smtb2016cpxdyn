### Orbits, fixed points and periodic points


 Consider the complex dynamical system in $\mathbb{C}$ defined by the holomorphic function $f(z)=z^2$. Our goal in this task is to study its orbits and the different possible behaviors.
- Define a function that returns the square of a complex number.
- Create a list containing the complex number 2i.
- Create a 'for' loop that calculates 5 iterations of the function starting with 2i, and adds each iteration to the list.
- Plot the list using the function provided.

```python
#Define the function
def square(z):
    return z**2
```

```python
#Create a list with 2i
iterates=[2j]
```

```python
#Compute the first 5 elements of the orbit
z = 2j
for i in range(5):
    z = square(z)
    iterates = iterates + [z]
    
print(iterates)
```

Now generalize it to define a function called $\tt iterate5square \tt (z)$ that takes a complex number $z$ as input, and returns a list with the first 5 elements of the orbit of that point

```python
#Define the function iterate5_square
def iterate5_square(z):
    #include the starting point 
    list_iter = [z]
    #loop 5 times
    for i in range(5):
        z = square(z)
        list_iter = list_iter+[z]
    return list_iter
```

### Plotting orbits of points

The function defined below plots a list of complex numbers

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_list (list_complex):
    width = 10
    height = 10
    plt.figure(figsize = (width, height))
    plt.plot([z.real for z in list_complex], [z.imag for z in list_complex],'ro')
    plt.axis([-4.1, 4.1, -4.1, 4.1])
```

You can use it in the following way to plot the first 5 iterates of $f(z)=z^2$ starting with $0.7-0.7i$ with the following command:

```python
plot_list(iterate5_square(0.7-0.7*1j))
```

### Example $g(z)=z^3-2iz^2$
Find the fixed points, classify them and use the functions that you have defined to plot some examples of orbits of points

```python
#Do your work here
def functiong(z):
    return z**3 - 2j*z**2

def iterate5_g(z):
    #include the starting point 
    list_iter = [z]
    #loop 5 times
    for i in range(5):
        z = functiong(z)
        list_iter = list_iter + [z]
    return list_iter
```

One can check that 0 is an attracting fixed point of the function and $i$ is a neutral fixed point. 

```python
def plot_list_size (list_complex,sz):
    width = 10
    height = 10
    plt.figure(figsize=(width, height))
    plt.plot([z.real for z in list_complex], [z.imag for z in list_complex],'ro')
    plt.plot([0],[0],'go')
    plt.axis([-sz, sz, -sz, sz])

plot_list_size(iterate5_g(0.1+0.1j),1)
plot_list_size(iterate5_g(0.2+0.2j),1)
plot_list_size(iterate5_g(0.3+0.3j),1)
plot_list_size(iterate5_g(0.4+0.4j),1)
plot_list_size(iterate5_g(0.5+0.5j),1)
```
